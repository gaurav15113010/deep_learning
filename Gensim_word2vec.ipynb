```python
import numpy
print("NumPy version:", numpy.__version__)

from gensim.models import Word2Vec
print("Gensim is ready!")
```

```python
with open("/01 Harry Potter and the Sorcerers Stone.txt","r") as f:
  text = f.read()
```

```python
text
```

```python
# preprocessing(we can do multiple preprocessing but here just for learning we are doing only 1 step of preprocessing  )
```

```python
import re
```

```python
processed_text = re.sub('[^A-Za-z. ]','',text)
```

```python
processed_text
```

```python
# word2vec require data in a specific format like below
#[[word1sent1,word2sent1],[word1sent2,word2sent2]] ...and so on
```

```python
text_list = processed_text.split('.')
```

```python
text_list[:10]
```

```python
# now we have a list of all the sentences, and now we want the list of all the words in a sentence
sentences_list = [sent.split() for sent in text_list]
```

```python
len(sentences_list)
```

```python
#creating model object
```

```python
#here sg parameter means skip gram , by default [sg=0, means,CBOW method] [sg=1, means skipgram method, this is 2nd architecture of sg]
#vector_size = dimension of the embedding required,(hidden layer neuron will be decided)
#window=5, this is by defaultn is 5, means how many words to be prdeicted in left and right
#min_count=1, 1 is by default, means, the minimum how many times the word should occur in the vocab (if min_count=10, means, all the words whose,
#     count is less than 10 will not be used for the processing, but since we have used 1, that means, even if any word has occured only once in
#     the vocab, still we will be considering that word for processing)
```

```python
model = Word2Vec(sentences=sentences_list,sg=1,vector_size=100,window=5,min_count=1)
```

```python
print(model.wv.index_to_key) # this is the vocab/unique words
```

```python
#now we can check the vectors of any word we want.
#vector of word 'Harry' is as below , len of this vector will be 100 , as we already defined while creating the model object
model.wv['Harry']
```

```python
len(model.wv['Harry'])
```

```python
# to check the most similar word
model.wv.most_similar('Harry')
```

```python
# now to save this model
model.save('harry_word2vec_100dim.model')
```

```python
#now we will check how can we create embeddings of new data from this same model,
# for that we will list down all the required parameters which we used while training this model
```

```python
# model = Word2Vec(sentences=sentences_list,sg=1,vector_size=100,window=5,min_count=1)
#these were all the parameters used

sentences(required):
  input data ,where each sentences represented as list of Words

vector size (default =100):
  dimensionality of vector

window (default = 5)
  maximum distance between the context word and focus word

min_count(default = 5)
  min freq of word

workers(default = 3)
  no of cpu cores to use for training

alpha (default = 0.025)
  initial learning rate , this starts reducing as we start the training, deep learning Concept

min_alpha(default = 0.001)
  learning rate will linearly drop to this value

epochs(default=5):
  no of iterations over the corpus/dataset

sg(default=0)
  sg=0 --> CBOW
  sg=1 --> skipgram
```

```python
prev_model = Word2Vec.load('/content/harry_word2vec_100dim.model')
```

```python
prev_model.wv['ron'] # we can see here , that vector of small case "ron" is not there in current model
```

```python
# now we will create a dummy dataset in which we will keep small case words,
# which we did not do in the preprocessing of previous model, and will train these new dataset with our
# already trained model
```

```python
new_sents = [['ron','harry','hermoine'],['ron','harry','hermoine']]
# we are adding three new words in the vocab/unique words
```

```python
# checking the length of vocab in prev model
len(prev_model.wv.index_to_key)
```

```python
# adding the new three words
prev_model.build_vocab(new_sents,update=True)
```

```python
# now we will check the length of the voacb after adding three new words
len(prev_model.wv.index_to_key)
```

```python
# so we can see that new three words are added in the vocab

prev_model.wv['ron']
# here we have got the embeddings before training but these are just randmoly intiliased numbers
```

```python
prev_model.train(new_sents,total_examples=prev_model.corpus_count,epochs=prev_model.epochs)
```

```python
prev_model.wv['ron']
# now these are the actual vectors, might be looking similar to the earlier
# but these is becuse we didnt trained on enough data to give model much context.
```

```python
prev_model.save("new_harry_word2vec_100dim.model")
```
